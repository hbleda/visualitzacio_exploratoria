```python
# Importamos las librerías necesarias para la realización del ejercicio.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.graph_objects as go


# Cargamos los datos del archivo csv en un dataframe y vemos la estructura de los primeros registros.
df_lloguer = pd.read_csv ("preu lloguer per trimestre i barri.csv")
df_lloguer.head()
```

```python
# Obtenemos información del tipo de cada uno de los campos del dataframe.
df_lloguer.info()
```

```python
# Comprobamos si tenemos valores nulos en el dataset
df_lloguer.isnull().sum()
```

```python
# convertimos el tipo de 'preu' de object a float, para poder tener un valor numérico
df_lloguer= df_lloguer['Preu'].astype(float)  
```

 >Por el error en la ejecución de la conversión, podemos comprobar que en el campo 'preu' tenemos algunos registros con 
 >valor '--'.

```python
#Contabilizamos el número de registros que tienen este valor
df_lloguer[df_lloguer['Preu'] =='--' ]['Nom_Districte'].count()
```

```python
#Comprobamos si hay alguna relación de los diferentes campos con los campos con  'Preu' con valor '--'.
df_lloguer[df_lloguer['Preu'] =='--'].sort_values(by='Nom_Barri')
```

>De la exploración visual de estos datos, podemos ver que los registros con 'Preu' == '--' corresponden a todos los registros de 5 barrios (Can Peguera, Torre Baró, Vallbona, la Clota, la Marina del Prat Vermell) y a los registros correspondientes a los trimestres 1 y 3 del barrio de Baró de Viver.

>En este punto, teniendo en cuenta lo que pide el ejercicio, tenemos una disyuntiva: para obtener valores numéricos basado en el campo 'Preu' es necesario convertir al tipo float este campo, pero para ello tenemos que eliminar los registros con el campo preu a '--', pero por otro lado al sacarlos estamos eliminando barrrios completos de un distrito, por lo que al sacar la información de los barrios existentes en cada distrito los datos no serán reales.

>Es por eso que en este punto extraigo la gráfica de barrios por distrito dentro de lo que solicita el primer punto del Ejercicio1 y posteriormente elimino los registros con 'Preu' == '--'.



## Una variable categòrica (Districte o Barri)

```python
# Número de barrios por distrito
ax= df_lloguer[['Nom_Districte','Nom_Barri']].drop_duplicates(subset = 'Nom_Barri').groupby("Nom_Districte")['Nom_Barri'].count().sort_values(ascending=False).plot(kind='barh',color='#1e9143',edgecolor='blue',width=0.6)
ax.bar_label(ax.containers[0], fontsize=10, color='black', label_type='edge')
ax.set_xticks([])
plt.xlabel('')
plt.ylabel('Nombre de los distritos')
plt.title('Número de Barrios por distrito')
# Guardar imagen
plt.savefig("Numero_barrios_distrito.jpg",bbox_inches='tight' )
```

>Podemos comprobar que el distrito con más barrios es el distrito de Nou Barris, con 13 y el que menos tiene,  el distrito de >Les Corts con 3, una cuarta parte de los que existen en Nou Barris.

```python
# limpiamos el dataframe de los valores '--' en 'Preu'
df_lloguer = df_lloguer.loc[df_lloguer['Preu'] != '--']
df_lloguer[df_lloguer['Preu'] =='--' ]['Nom_Districte'].count()
```

```python
# Ahora si podemos realizar la conversión de tipo en el campo 'Preu'
df_lloguer['Preu']= df_lloguer['Preu'].astype(float)
df_lloguer.info()
```

## Una variable numèrica (Preu)

```python
# Obtenemos la gráfica de la distribución del precio medio del m2/mes en los diferentes distritos. La figura del violin
# nos muestra de forma interactiva los valores que obtenemos con la función .describe() y además podemos ver el resto de
# valores y su kde(kernel density estimation).

fig = go.Figure(data=go.Violin(y=df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)']['Preu'], box_visible=True, line_color='black',
                               meanline_visible=True, fillcolor='#1e9143', opacity=0.8,
                               x0='Precio m2/mes'))

fig.update_layout(yaxis_zeroline=False)
fig.show()
fig.write_html("C:\\Users\\USER\Desktop\Data Science\\Sprint 4. Visualització gràfica de dades\\Tasca M4 T01\\Precio_m2_mes.html")
```

> De la gráfica podemos observar de forma visual e interactiva, que el precio mínimo por m2 corresponde a 6.70 euros m2/mes y el máximo 18 euros m2/mes.La media es de 12.72 euros m2/mes. El valor de kde alcanza su densidad máxima en los 12,95 m2/mes, lo que nos indica que la probabilidad de que un nuevo valor este próximo a este precio es muy alta.

```python
df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)']['Preu'].describe()
```

## Una variable numèrica i una categòrica (Districte i Preu)

```python
# En este caso utilizaremos el precio de alquiler mensual para realizar la gráfica
plt.figure(figsize=(25,10))
g=sns.boxplot(x='Nom_Districte', y='Preu', data=df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)'], palette= 'rainbow')
g.set_title('Precio de alquiler mensual por Distrito', fontsize=25)
g.set_xlabel('Distritos',fontsize=20)
g.set_ylabel('Euros/Mes', fontsize=20)
g.tick_params(labelsize=15)
# Guardar imagen
plt.savefig("Precio_medio_mes_por_distrito.jpg",bbox_inches='tight')
```

> En este caso utilizamos el precio mensual y lo distribuimos por los diferentes distritos de la ciudad. Podemos ver 
que los valores más altos corresponden al distrito de Sarrià-Sant Gervasi y los más bajos al distrito de Nou Barris 
y Sant  Andreu(debemos tener en cuenta que dentro de estos datos faltan 3 barrios de Nou Barris y 1 de Sant Andreu, 
pero que posiblemente no hubieran modificado la distribución).         

## Dues variables numèriques (Any o Trimestre i Preu)

```python
plt.plot(df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)'].groupby('Trimestre')['Preu'].mean(),marker = 'o', ms = 20,color='#1e9143',mec = 'yellow')
plt.xticks([1, 2, 3, 4], ['1', '2', '3', '4'])
plt.xlabel('Trimestre')
plt.ylabel('Precio euros m2/es')
plt.grid(True)
plt.title('Precio euros m2/es Barcelona 2021')

# Guardar imagen
plt.savefig("Precio_euros_m2_mes.jpg",bbox_inches='tight' )
```

> Según esta distribución del precio por m2/mes en la ciudad de Barcelona, parece que la mejor época para buscar un alquiler en el año 2021 fue durante los meses de abril, mayo y junio (segundo trimestre), donde el precio alcanza su nivel más bajo quedando un poco por encima de los 12,55 euros m2/mes. Todos aquellos que buscaron un alquiler en la parte final del año tuvieron los precios más caros de media.

## Tres variables (Barri o Districte, Trimestre i Preu)

```python
k =sns.relplot(data=df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)'].pivot_table('Preu', 'Trimestre', 'Nom_Districte'), kind='line',linewidth = 3, height=10)
k._legend.remove()
plt.ylabel('Euros m2/mes', fontsize=20)
plt.xlabel('Trimestre', fontsize=20)
plt.tick_params(labelsize=17)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.,title='Distritos', fontsize='14', title_fontsize='18')

# Guardar imagen
plt.savefig('Precio_trimestre_m2_mes_distrito.jpg', bbox_inches='tight')
```

> Siguiendo con el comentario de la gráfica anterior, podemos decir que la bajada en el precio medio del m2/mes en la ciudad de Barcelona durante el año 2021 viene motivada especialmente por la baja del precio en el distrito de Sant Andreu durante esa época del año, siendo incluso inferior este precio al del distrito de Nou Barrís que tiene precios más bajos a lo largo de todo el año.  Esta gráfica es interesante, ya que podemos ver que a finales de año en todos los distritos hay una tendencia al alza del precio más o menos moderada (siendo muy alta en el distrito de Sarrià-Sant Gervasi), excepto en los distritos de Sant Martí y Les Corts.

```python
plt.figure(figsize=(12,6))
df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà mensual (Euros/mes)'].groupby(['Nom_Barri']).max('Preu').sort_values('Preu', ascending=False)['Preu'].plot.bar()
plt.title('Precio Máximo por mes por Barrio')
plt.tick_params(labelsize=10)
plt.xlabel('Barrios')
plt.ylabel('Precio Máximo por mes en Euros')
# Guardar imagen
plt.savefig('Precio_maximo_mes_barrio.jpg', bbox_inches='tight')
```

```python
plt.figure(figsize=(12,6))
df_lloguer.loc[df_lloguer.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)'].groupby(['Nom_Barri']).max('Preu').sort_values('Preu', ascending=False)['Preu'].plot.bar()
plt.title('Precio Máximo por m2/mes por Barrio')
plt.tick_params(labelsize=10)
plt.xlabel('Barrios')
plt.ylabel('Precio Máximo por m2/mes en Euros')
# Guardar imagen
plt.savefig('Precio_maximo_m2_mes_barrio.jpg', bbox_inches='tight')
```

> En este caso las 2 gráficas nos muestran los precios máximos por barrio, tanto en precio por mes como en precio por m2/mes. Mientras que en la primera los precios máximos los encontramos en el barrio de Pedralbes, en la gráfica del precio máximo m2/mes este barrio lo encontramos en la septima posición.

> Las conclusiones, y sin disponer de la información de la media de los m2 por vivienda por cada uno de los barrios, es que es posible que esta media sea más alta en el barrio de Pedralbes que en el barrio de Vallvidrera, el Tibidabo i les Planes.
